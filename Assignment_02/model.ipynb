# Tạo mô hình để dự đoán khi nào nhân viên sẽ ra đi

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.describe(include='O')
```

bỏ cột Over18 vì feature này không có gì để cho vào mô hình

```python
df = df.drop(columns=["Over18"])
```

```python
df["Attrition"].value_counts() /len(df)
```

**Thử với label encode**

```python
le = LabelEncoder()
n = ["Attrition", "BusinessTravel", "Department", "EducationField", "Gender", 
             "JobRole", "MaritalStatus", "OverTime"]
for i in n:
    df[i] = le.fit_transform(df[i])
y = df["Attrition"]
X = df.drop(columns=["Attrition"])
```

```python
#X.shape
```

**Holdout**

```python
from sklearn.model_selection import train_test_split
(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

**Build model**

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

**Evaluate model**

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
y_pred = model.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred))
```
